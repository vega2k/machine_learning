```python
#광역시도 컬럼 값을 선택 (unique)
data['광역시도'].unique()
```

```python
#경기도의 인구수의 평균과 표준편차
#Boolean indexing
ky = data['광역시도'] == '경기도'
```

```python
# loc은 행과 열을 선택할때 사용
data.loc[ky,['인구수','광역시도','행정구역']].head()
```

```python
#전국 인구수 평균
print('전국', data['인구수'].mean())
#경기도 인구수 평균
print('경기도', data.loc[ky,'인구수'].mean())
#경기도 인구수 표준편차
print('경기도', data.loc[ky,'인구수'].std())
```

### Data Selection
* loc[], iloc[] 을 사용
  - Slicing 사용
  - 특정행과 열을 여러개 선택 [] 사용
  - 조건식으로 특정행을 선택

```python
# 모든 행과 인구수,행정구역 2개의 컬럼 선택
data.loc[0:10:3,['인구수','행정구역']]
```

```python
# 인덱스가 0,3,5 , 인구수 부터 광역시도 컬럼까지
data.loc[[0,3,5],'인구수':'행정구역']
```

```python
#광역시도가 서울특별시 인 행과 열을 선택하기
#인구수 desc, index 초기화
seoul_df = data.loc[data['광역시도'] == '서울특별시']\
.sort_values(by='인구수',ascending=False).reset_index(drop=True)
seoul_df
```
