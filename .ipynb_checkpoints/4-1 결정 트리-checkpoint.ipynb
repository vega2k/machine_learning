```python
#독립변수
#훈련데이터 / 테스트데이터
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()

#종속변수
#훈련데이터 / 테스트데이터
target = wine['class'].to_numpy()

print(type(data))
print(type(target))
```

```python
#X_train : 훈련데이터 독립변수
#y_train : 훈련데이터 종속변수
#X_test : 테스트데이터 독립변수
#y_test : 테스트데이터 종속변수
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#feature들의 평균값
print(X_train.mean(axis=0))
#feature들의 분산값
X_train.var(axis=0)
```

 * StandardScaler는 평균과 표준편차를 사용해서 데이터 분산을 평준화 하는 방법
 * 평균(mean)은 주어진 수의 합을 측정개수로 나눈값
 * 편차는 변량 - 평균
 * 분산(variance)은 변량들이 퍼져있는 정도, 분산이 크면 데이터가 고르지 않다는 의미
  - 편차제곱의 평균
 * 표준편차(standard deviation) : 분산은 수치가 너무 커서, 제곱근으로 적당하게 줄인값
 * [표준편차설명](https://learnx.tistory.com/entry/%ED%86%B5%EA%B3%84%EC%9D%98-%EA%B8%B0%EC%B4%88%EC%9D%B8-%ED%8F%89%EA%B7%A0-%EB%B6%84%EC%82%B0-%ED%91%9C%EC%A4%80%ED%8E%B8%EC%B0%A8)

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(X_train)

train_scaled = ss.transform(X_train)
test_scaled = ss.transform(X_test)
```

```python
#feature들의 평균값
print(train_scaled.mean(axis=0))
#feature들의 분산값
train_scaled.var(axis=0)
```

## 결정 트리
* 결정트리 모델은 데이터를 잘 나눌 수 있는 질문을 찾는다면 계속 질문을 추가해서 분류 정확도를 높일 수 있다


* criterion : 분할 품질을 측정하는 기능 (default : gini)
* splitter : 각 노드에서 분할을 선택하는 데 사용되는 전략 (default : best)
* max_depth : 트리의 최대 깊이 (값이 클수록 모델의 복잡도가 올라간다.)
* min_samples_split : 자식 노드를 분할하는데 필요한 최소 샘플 수 (default : 2)
* min_samples_leaf : 리프 노드에 있어야 할 최소 샘플 수 (default : 1)
* min_weight_fraction_leaf : min_sample_leaf와 같지만 가중치가 부여된 샘플 수에서의 비율
* max_features : 각 노드에서 분할에 사용할 특징의 최대 수
* random_state : 난수 seed 설정
* max_leaf_nodes : 리프 노드의 최대수
* min_impurity_decrease : 최소 불순도
* min_impurity_split : 나무 성장을 멈추기 위한 임계치
* class_weight : 클래스 가중치
* presort : 데이터 정렬 필요 여부

### 가지치기
* 트리의 최대깊이를 지정한다 
* max_depth 매개변수를 3으로 지정한다

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(train_scaled, y_train)

print(dt.score(train_scaled, y_train))
print(dt.score(test_scaled, y_test))
```

* 깊이1의 노드는 모두 당도(sugar)는 기준으로 훈련세트를 나눈다
* 깊이2의 노드는 맨 왼쪽의 노드만 당도를 기준으로 나누고 두번째 노든 알코올도수(alcohol)를 기준으로 나누고, 오른쪽의 두 노드는 ph를 사용한다
* 깊이3에 있는 노드가 최종노드이다 왼쪽에서 세번째는 음성이 더 많고, 이노드에 도착해야 레드와인으로 예측한다
* 이 노드까지 도달하려면 당도는 -0.239 보다 작고, -0.802보다 작아야 합니다.알콜도수는 0.454보다 작아야 합니다 
* 즉 당도가 -8.802와 같거나 알코올도수가 0.454와 같거나 작은 것이 레드와인입니다. 
* -0.802 라는 음수로 된 당도를 어떻게 설명할 것인가?
* 특성값의 스케일은 결정트리 알고리즘에 아무런 영햐을 미치지 않으므로 표준화 전처리를 할 필요가 없다
* 이것이 결정트리알고리즘의 장점이다

## 확인문제
* min_impurity_decrease를 사용해서 가지치기를 해도 된다

```python
dt = DecisionTreeClassifier(min_impurity_decrease=0.0005, random_state=42)
dt.fit(X_train, y_train)

print(dt.score(X_train, y_train))
print(dt.score(X_test, y_test))
```
