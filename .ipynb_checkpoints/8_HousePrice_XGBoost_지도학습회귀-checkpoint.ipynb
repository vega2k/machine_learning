### 1. Load Data
* 훈련데이터와 테스트데이터 읽어오기


```python
import pandas as pd
import os

data_path = '/content/drive/My Drive/Colab Notebooks/data/house'
print(data_path)

train = pd.read_csv(os.path.join(data_path, 'train.csv'))
print(train.shape)

test = pd.read_csv(os.path.join(data_path, 'test.csv'))
print(test.shape)
```

```python
print(train.columns)
train.head()
```

```python
# Id 컬럼을 index로 변경하기
# inplace = True는 원본 데이터에 즉시 반영한다
train.set_index('Id', inplace=True)
test.set_index('Id', inplace=True)

print(train.shape, test.shape)

train.head()
```

### 2. Feature Selection
* 독립변수와 종속변수 간의 상관관계(Correlation) 분석
* corr() 함수로 변수간의 상관관계를 담은 데이터프레임을 생성 ( 상관계수값이 0.3 이상)
* 상관계수는 -1 ~ 1 사이의 값, 1에 가까울 수록 관련성이 있음
* Seaborn의 Heatmap 으로 시각화

```python
corr_mat = train.corr()
corr_mat.head(2)
```

```python
#corr_mat.index
top_corr_features = corr_mat.index[abs(corr_mat['SalePrice']) >= 0.3]
top_corr_features
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(13,10))
sns.heatmap(train[top_corr_features].corr(), annot=True, cmap="RdYlGn")
```

### 2. Data 전처리
* 수치형컬럼은 null 건수가 50 % 이상인 경우는 삭제
 - 남은 null data는 평균값을 채워주기
 - scikit-learn에서 제공하는 SimpleImputer를 사용
* train와 test 합치기
 - concat() 사용
* 범주형컬럼과 수치형컬럼을 나누기
* 범주형컬럼(object)은 전부 One Hot Encoding 
 - get_dummies() 사용
* 범주형데이터와 수치형데이터를 합치기
 - merge() 사용



```python
# SalePrice 컬럼값을 삭제하기 전에 저장
train_y_label = train['SalePrice']
print(train_y_label.shape)

train.drop(['SalePrice'], axis='columns', inplace=True)
print(train.shape, test.shape)
```

```python
# train와 test 합치기
boston_df = pd.concat((train, test), axis=0)
print(boston_df.shape)
boston_df_index = boston_df.index
```

```python
boston_df.tail(3)
```

```python
boston_df_index
```

```python
boston_df.isnull().sum().sort_values(ascending=False)
```

```python
check_null_gt5 = boston_df.isnull().sum() / len(boston_df)
check_null_gt5
```

```python
check_null_gt5[check_null_gt5 >= 0.5]
```

```python
drop_col_names = check_null_gt5[check_null_gt5 >= 0.5].index
drop_col_names
```

```python
# null값이 50% 이상인 컬럼삭제
boston_df.drop(drop_col_names,axis='columns', inplace=True)
boston_df.shape
```

```python
# 범주형 과 수치형 으로 데이터 분리 
# 범주형 컬럼 39, 수치형 36
# 범주형
boston_obj_df = boston_df.select_dtypes(include='object')
# 수치형
boston_num_df = boston_df.select_dtypes(exclude='object')

print(boston_obj_df.shape, boston_num_df.shape)
```

```python
boston_obj_df.columns
```

```python
boston_num_df.columns
```

```python
boston_obj_df.isnull().sum()
```

```python
# One Hot Encoding
boston_dummy_df = pd.get_dummies(boston_obj_df, drop_first=True)

print(boston_dummy_df.shape)
```

```python
boston_dummy_df.head(2)
```

```python
# 수치형 데이터 null count확인
boston_num_df.isna().sum().sort_values(ascending=False)
```

```python
# scikit-learn 에서 제공하는 SimpleImputer를 사용하여 null 값을 평균값으로 치환하기
import numpy as np
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
# fit
imputer.fit(boston_num_df)
# tranform
boston_num_df_ = imputer.transform(boston_num_df)
print(type(boston_num_df_))
```

```python
boston_num_df_[:2]
```

```python
# 평균으로 치환된 값만 들어있는 2차원배열을 DataFrame으로 생성하기
boston_num_df_notnull = pd.DataFrame(boston_num_df_, columns=boston_num_df.columns, index=boston_df_index)
print(boston_num_df_notnull.shape)
```

```python
boston_num_df_notnull.isnull().sum()
```

```python
# boston_dummy_df 와 boston_num_df_notnull 을 merge
print(boston_dummy_df.shape, boston_num_df_notnull.shape)
```

```python
boston_dummy_df.index
```

```python
boston_num_df_notnull.index
```

```python
# merge 할때 index 순서대로 맞추기 위해서 왼쪽과 오른쪽 index를 True로 지정해야 한다
boston_df = pd.merge(boston_dummy_df, boston_num_df_notnull, left_index=True, right_index=True )
print(boston_df.shape)
boston_df.info()
```

```python
train_length = len(train)

# SettingWithCopyWarning: 원천 DataFrame과 별도의 메모리 공간을 갖는 파생된 DataFrame을 생성해라는 warning
# warning을 방지하려면 copy()로 복사를 해서 사용하는 것이 좋다
train_df = boston_df[:train_length].copy()
test_df = boston_df[train_length:].copy()

# SalePrice 컬럼추가
train_df['SalePrice'] = train_y_label

print(train_df.shape, test_df.shape)
```
